Импортируем модули для работы с данными

```python
import requests
import pandas as pd
import psycopg2
```

Базовая часть ссылки

```python
api_base_url = 'https://www.anapioficeandfire.com/api'
```

## Книги

```python
books_list = requests.get(
    f'{api_base_url}/books/'
).json()

print(f'Всего написано книг - {len(books_list)}')
```

Преобразование в Pandas DataFrame

```python
df_books = pd.DataFrame(books_list)
df_books.head()
```

## Дома Вестероса

```python
houses_list = requests.get(
    f'{api_base_url}/houses/'
).json()

print(f'Всего домов - {len(houses_list)}')
```

Преобразование в Pandas DataFrame

```python
df_houses = pd.DataFrame(houses_list)
df_houses.head()
```

## Дома Вестероса с девизом

```python
houses_with_title_list = requests.get(
    f'{api_base_url}/houses?hasTitles=True'
).json()

print(f'Всего домов с девизом - {len(houses_with_title_list)}')
```

Преобразование в Pandas DataFrame

```python
df_houses_with_title = pd.DataFrame(houses_with_title_list)
df_houses_with_title.head()
```
