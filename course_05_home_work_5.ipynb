# Оптимизация типов pandas

## Считаем датасет из файла

```python
df_house = pd.read_csv('Housing.csv')
```

```python
df_house.head()
```

## Для каждого столбца определим оптимальный с точки зрения потребления памяти тип данных

```python
df_house.info()
```

```python
df_house.describe()
```

```python
numeric_cols = df_house.select_dtypes(include=['number']).columns
print(numeric_cols)
```

```python
df_house = df_house.fillna(0)
```

```python
def int8_type(number):
    if -128 <= number <= 127:
        return True
    return False


def int16_type(number):
    if -32768 <= number <= 32767:
        return True
    return False


def int32_type(number):
    if -2.1*10**9 <= number <= 2.1*10**9:
        return True
    return False


def int64_type(number):
    if -9.2*10**18 <= number <= 9.2*10**18:
        return True
    return False


def cast_numeric_column(df_col):
    max = df_col.max()
    min = df_col.min()
    if(int8_type(min) and int8_type(max)):
        return df_col.astype(np.int8)
    if(int16_type(min) and int16_type(max)):
        return df_col.astype(np.int16)
    if(int32_type(min) and int32_type(max)):
        return df_col.astype(np.int32)
    if(int64_type(min) and int64_type(max)):
        return df_col.astype(np.int64)


for col_name in numeric_cols:
   df_house[col_name] = cast_numeric_column(df_house[col_name])
```

```python
df_house.info()
```

После анализа числовых колонок пришел к выводам: 1) Значения колонок 'bedrooms', 'bathrooms', 'stories', 'parking' по диапазону подходили под тип int8; 2) Колонка 'area' уже не помещалась в int8, но поместилась в диапазон int16; 3) Самая большая по диапазону стала колонка 'price' которая подошла по значениям к типу int32. Преобраование типов позволило уменьшить потребление оперативной памяти

```python
non_numeric_cols = df_house.select_dtypes(exclude=['number']).columns
print(non_numeric_cols)
```

```python
def cust_to_category_object(df_col):
    print('Информация по занимаемой памяти до преобразования типа - \n')
    print(df_house[df_col].info(memory_usage="deep"))
    print('')
    print('Информация по уникальным значениям в колонке - \n')
    print(df_house[df_col].unique())
    df_house[df_col] = df_house[df_col].astype("category")
    print('')
    print('Информация по занимаемой памяти после преобразования типа - \n')
    print(df_house[df_col].info(memory_usage="deep"))


for col_name in non_numeric_cols:
    cust_to_category_object(col_name)
```

```python
df_house.info()
```

По итогу удалось уменьшить в четыре раза потребления оперативной памяти, с 55.5 КБ в начале до 10.0 КБ в конце за счет преобразования типов
