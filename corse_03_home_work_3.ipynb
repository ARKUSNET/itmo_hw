 ## Выведем самое популярное мужское и самое популярное женское имя на корабле

Объявим функцию по поиску и извлечению женского имени с использованием регулярного выражения (в функции используется второй поиск для тех случаев когда замужняя девушка имеет имя мужа и в круглых скобках своё имя, например - Futrelle, Mrs. Jacques Heath (Lily May Peel))

```python
def parse_first_name_female(name):
    first = name.str.extract(r"Mrs\.\s+[^(]*\((\w+)", expand=False)
    first.loc[first.isna()] = name.str.extract(r"\.\s+(\w+)", expand=False)
    return first
```

Объявим функцию по поиску и извлечению мужского имени 

```python
def parse_first_name_male(name):
    first = name.str.extract(r"(Mr\.|Dr\.|Rev\.|Master\.|Don\.|Major\.|Sir\.|Col\.|Capt\.|Jonkheer\.) ([a-zA-Z]+)", expand=False)[1]
    return first
```

```python
male_names = parse_first_name_male(df.loc[df["Sex"] == "male", "Name"])
male_names.value_counts(ascending=False).head(1)
```

```python
femaly_names = parse_first_name_female(df.loc[df["Sex"] == "female", "Name"])
femaly_names.value_counts().head(1)
```

По итогу самое частое имя среди мужчин - William, а среди женщин - Anna.

 ## Выведем количества n-местных кабин (в которых было 2, 3, 4, ... человека)

Заполним пустые поля в номере кабины значением - "Cabin missing"

```python
df["Cabin"] = df["Cabin"].apply(lambda x: "Cabin missing" if pd.isnull(x) else x)
```

```python
df["Ticket"].value_counts()
```

```python
df[df["Ticket"] == '347082']
```

Это были пассажиры, которые были семьей, к сожалению в примере никто не выжил

Зададим две функции по определению вместимости кабины и номера класса

```python
def get_cabin_capacity(cabin_obj):
    for key, value in cabin_obj.to_dict().items():
        if(value > 1):
            print(f'{value} - местная кабина № {key}, находится в {get_class_number_by_cabin(key)} классе.')

        
def get_class_number_by_cabin(cabin):
    return df.loc[df["Cabin"] == cabin, "Pclass"].iloc[0]


get_cabin_capacity(df["Cabin"].value_counts())
```

По итогу: в основном двух, трёх и четырех местные кабины располагались в первом и во втором классе. Для 687 пассажиров не был определен номер кабины в третьем классе, возможно пассажиры находились в холле и/или каюты не были достроены.
