Создаем класс

```python
class Chocolate:
    def __init__(self, price, additive = ''):
        """
        Метод инициализации для класса Chocolate
        :param self: ссылка на этот объект
        :param price: цена шоколада, тип float
        :param additive: добавки в шоколад, тип str, опциональный
        """
        if price < 0:
            raise ValueError('Цена должна быть больше нуля!')
        self.price = price
        self.additive = additive

    def show_my_chocolate(self):
        """
        Метод выводит информацию о шоколаде
        :param self: ссылка на этот объект
        """
        if self.additive != '':
            return f'Шоколадный батончик, добавка: {self.additive}'
        else:
            return 'Обычный шоколадный батончик'

    def __str__(self):
        """
        Метод выводит информацию об объекте Chocolate, переопределяет метод print()
        :param self: ссылка на этот объект
        """
        return f'Шоколадный батончик, {self.price} рублей'
```

Проинициализируем класс Chocolate с одним аргументом - цена (больше нуля)

```python
chocolate = Chocolate(200.123)
print(chocolate.show_my_chocolate())
print(chocolate)
```

Проинициализируем класс Chocolate c двумя аргументами - price (больше нуля), additive

```python
chocolate = Chocolate(100.234, 'Lemon')
print(chocolate.show_my_chocolate())
print(chocolate)
```

Проинициализируем класс Chocolate с одним аргументом - price (меньше нуля)

```python
chocolate = Chocolate(-100.09)
print(chocolate.show_my_chocolate())
print(chocolate)
```
